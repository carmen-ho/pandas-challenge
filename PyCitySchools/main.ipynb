```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
schools_file = "Resources/schools_data.csv"
students_file = "Resources/students_data.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
schools_info = pd.read_csv(schools_file)
students_info = pd.read_csv(students_file)

schools_info
```

```python
students_info
```

```python
# Combine the data into a single dataset.  
schools_analysis = pd.merge(students_info, schools_info, how="left", on=["school_name"])
schools_analysis
```

```python
# Calculate the total number of schools

unique_schools_name=schools_analysis["school_name"].unique()

total_schools_number=len(unique_schools_name)

total_schools_number
```

```python
# Calculate the total number of students

total_students_number=schools_analysis["student_name"].count()

total_students_number
```

```python
# Calculate the total budget

total_budget = schools_info["budget"].sum()

total_budget
```

```python
# Calculate the average math score

average_math_score = schools_analysis["math_score"].mean()

average_math_score
```

```python
# Calculate the average reading score

average_reading_score = schools_analysis["reading_score"].mean()

average_reading_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)

students_passing_math = students_info[schools_analysis["math_score"] >= 70 ].count()["Student ID"]

percent_passing_math = (students_passing_math / total_students_number) * 100

percent_passing_math
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)

students_passing_reading = students_info[schools_analysis["reading_score"] >= 70 ].count()["Student ID"]

percent_passing_reading = (students_passing_reading / total_students_number) * 100

percent_passing_reading
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)

overall_passing = schools_analysis[(schools_analysis['math_score'] >= 70) & (schools_analysis['reading_score'] >= 70)]['Student ID'].count()/total_students_number*100

overall_passing
```

```python
# Create a dataframe to hold the above results

district_summary = pd.DataFrame({
    "Total Schools": total_schools_number,
    "Total Students": f"{total_students_number:,}",
    "Total Budget": f"${total_budget:,.2f}",
    "Average Math Score": f"{average_math_score:.6f}",
    "Average Reading Score": f"{average_reading_score:.5f}",
    "% Passing Math": f"{percent_passing_math:.6f}",
    "% Passing Reading": f"{percent_passing_reading:.6f}",
    "% Overall Passing": f"{overall_passing: .6f}"
}, index=[0])

district_summary
```

```python
# School Type (by school name)

schools_type = schools_info.set_index('school_name')['type']

schools_type
```

```python
# Total Students 

total_students = schools_analysis['school_name'].value_counts()

total_students
```

```python
# Total School Budget

total_schools_budget = schools_analysis.groupby(['school_name'])['budget'].mean()

total_schools_budget
```

```python
# Per Student Budget

budget_per_student = total_schools_budget / total_students

budget_per_student
```

```python
# Average Math Score

average_math_score = schools_analysis.groupby(['school_name'])['math_score'].mean()

average_math_score
```

```python
# Average Reading Score

average_reading_score = schools_analysis.groupby(['school_name'])['reading_score'].mean()

average_reading_score
```

```python
# % Passing Math

pass_math_percent = schools_analysis[schools_analysis['math_score'] >= 70].groupby('school_name')['Student ID'].count()/total_students*100

pass_math_percent
```

```python
# % Passing Reading

pass_read_percent = schools_analysis[schools_analysis['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/total_students*100

pass_read_percent
```

```python
# % Overall Passing (The percentage of students that passed math and reading)

overall_pass = schools_analysis[(schools_analysis['reading_score'] >= 70) & (schools_analysis['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/total_students*100

overall_pass
```

```python
# Create a dataframe to hold the above results

school_summary = pd.DataFrame({
    "Schools Type": schools_type,
    "Total Students": total_students,
    "Total School Budget": total_schools_budget,
    "Per Student Budget": budget_per_student,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": pass_math_percent,
    "% Passing Reading": pass_read_percent,
    "% Overall Passing": overall_pass
})

school_summary.style.format({'Total Students': '{:}',
                          "Total School Budget": "${:,.2f}",
                          "Per Student Budget": "${:.2f}",
                          'Average Math Score': "{:6f}", 
                          'Average Reading Score': "{:6f}", 
                          "% Passing Math": "{:6f}", 
                          "% Passing Reading": "{:6f}"})

school_summary
```

```python
# Sort and display the top five performing schools by % overall passing

top_perform = school_summary.sort_values("% Overall Passing", ascending = False)

top_perform.head().style.format({'Total Students': '{:}',
                           "Total Schools Budget": "${:,.2f}", 
                           "Per Student Budget": "${:.2f}", 
                           "% Passing Math": "{:6f}", 
                           "% Passing Reading": "{:6f}", 
                           "% Overall Passing": "{:6f}"})
```

```python
# Sort and display the five worst-performing schools by % overall passing

bottom_perform = school_summary.sort_values("% Overall Passing", ascending = True)

bottom_perform.style.format({'Total Students': '{: }', 
                       "Total Schools Budget": "${:,.2f}", 
                       "Per Student Budget": "${:.2f}", 
                       "% Passing Math": "{:6f}", 
                       "% Passing Reading": "{:6f}", 
                       "% Overall Passing": "{:6f}"})

bottom_perform.head(5)
```

```python
# Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school

ninth_math = students_info.loc[students_info['grade'] == '9th'].groupby('school_name')["math_score"].mean()

tenth_math = students_info.loc[students_info['grade'] == '10th'].groupby('school_name')["math_score"].mean()

eleventh_math = students_info.loc[students_info['grade'] == '11th'].groupby('school_name')["math_score"].mean()

twelfth_math = students_info.loc[students_info['grade'] == '12th'].groupby('school_name')["math_score"].mean()


math_scores = pd.DataFrame({
        "9th": ninth_math,
        "10th": tenth_math,
        "11th": eleventh_math,
        "12th": twelfth_math
})

math_scores = math_scores[['9th', '10th', '11th', '12th']]

math_scores.index.name = "School Name (Math)"

math_scores.style.format({'9th': '{:.6f}', 
                          "10th": '{:.6f}', 
                          "11th": "{:.6f}", 
                          "12th": "{:.6f}"})
```

```python
# Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school

ninth_reading = students_info.loc[students_info['grade'] == '9th'].groupby('school_name')["reading_score"].mean()

tenth_reading = students_info.loc[students_info['grade'] == '10th'].groupby('school_name')["reading_score"].mean()

eleventh_reading = students_info.loc[students_info['grade'] == '11th'].groupby('school_name')["reading_score"].mean()

twelfth_reading = students_info.loc[students_info['grade'] == '12th'].groupby('school_name')["reading_score"].mean()


reading_scores = pd.DataFrame({
        "9th": ninth_reading,
        "10th": tenth_reading,
        "11th": eleventh_reading,
        "12th": twelfth_reading
})

reading_scores = reading_scores[['9th', '10th', '11th', '12th']]

reading_scores.index.name = "School Name (Reading)"

reading_scores.style.format({"9th": "{:.6f}", 
                             "10th": "{:.6f}", 
                             "11th": "{:.6f}", 
                             "12th": "{:.6f}"})
```

```python
# Create a table that breaks down school performances based on average Spending Ranges

bins = [0, 585, 630, 645, 680]
group_name = ["<$585", "$585-630", "$630-645", "$645-680"]

spending_df = school_summary

spending_df['Spending Ranges (Per Student)'] = pd.cut(budget_per_student, bins, labels = group_name, right=False)

spending_df
```

```python


avg_math = spending_df.groupby(['Spending Ranges (Per Student)'])['Average Math Score'].mean()

avg_read = spending_df.groupby(['Spending Ranges (Per Student)'])['Average Reading Score'].mean()

pass_math = spending_df.groupby(['Spending Ranges (Per Student)'])['% Passing Math'].mean()

pass_read = spending_df.groupby(['Spending Ranges (Per Student)'])['% Passing Reading'].mean()

overall = spending_df.groupby(['Spending Ranges (Per Student)'])['% Overall Passing'].mean()


scores_by_spend = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "% Passing Math": pass_math,
    "% Passing Reading": pass_read,
    "% Overall Passing": overall
})

scores_by_spend = scores_by_spend[[
    "Average Math Score",
    "Average Reading Score",
    "% Passing Math",
    "% Passing Reading",
    "% Overall Passing"
]]

scores_by_spend.style.format({"Average Math Score": "{:.2f}", 
                              "Average Reading Score": "{:.2f}", 
                              "% Passing Math": "{:.2f}", 
                              "% Passing Reading":"{:.2f}", 
                              "% Overall Passing": "{:.2f}"})

scores_by_spend
```

```python
# Create a table that breaks down school performances based on school size

bins = [0, 1000, 1999,5000]
group_name = ["Small (<1000)", "Medium (1000-2000)" , "Large (2000-5000)"]

school_size_df = school_summary

school_size_df['School Size'] = pd.cut(school_summary['Total Students'], bins, labels = group_name)

school_size_df
```

```python

average_math_score = school_size_df.groupby(['School Size'])['Average Math Score'].mean()

average_reading_score = school_size_df.groupby(['School Size'])['Average Reading Score'].mean()

pass_math_percent = school_size_df.groupby(['School Size'])['% Passing Math'].mean()

pass_read_percent = school_size_df.groupby(['School Size'])['% Passing Reading'].mean()

overall = school_size_df.groupby(['School Size'])['% Overall Passing'].mean()

         
scores_by_size = pd.DataFrame({
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    '% Passing Math': pass_math_percent,
    '% Passing Reading': pass_read_percent,
    '% Overall Passing': overall
            
})


scores_by_size = scores_by_size[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    '% Overall Passing'
]]

scores_by_size.style.format({'Average Math Score': '{:.6f}', 
                              'Average Reading Score': '{:.6f}', 
                              '% Passing Math': '{:.6f}', 
                              '% Passing Reading':'{:.6f}', 
                              '% Overall Passing': '{:.6f}'})
```
